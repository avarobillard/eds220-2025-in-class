# `geopandas` 

GeoPandas is a Python library that extends `pandas` by adding support for geospatial data

## Reproducible file paths

```python
import os
import geopandas as gpd
import matplotlib.pyplot as plt
```

To import data, we first use `os` package to create a reproducible file path

```python
fp = os.path.join('data', 'gbif_sus_scrofa_california', 'gbif_sus_scrofa_california.shp')
fp
```

```python
pigs = gpd.read_file(fp)
pigs.head()
```

```python
fp = os.path.join('data', 'ca_state_boundary', 'ca_state_boundary.shp')
ca_boundary = gpd.read_file(fp)
ca_boundary
```

## `Geoseries` and `GeoDataFrame` 

`geopandas.GeoDataFrame` = core structure in geopandas = `pandas.DataFrame` plus a dedicated geometry column that can perform spatial operations

**geometry column** hold the geometry (points, polygons, etc) of each spatial feature. The type of this column is `geopandas.GeoSeries`

```python
# Check the data type of the pigs dataframe
print(type(pigs))

# Check the data type of the geometry column 
print(type(pigs.geometry))

# Check data type of gbifID column
print(type(pigs.gbifID))
```

```python
# Check data type of each column
pigs.dtypes
```

```python
# Check type of each element in geometry column
pigs.geom_type
```

What is the geometry type of the single feature in CA state boundary?

```python
ca_boundary.geom_type
```

## CRS

**Coordinate reference system**- instructions to locate each spatial feature of our data frame on the surface of the Earth

```python
# Access the CRS of the GeoDataFrame
pigs.crs
```

```python
# Examine CRS details
print('Ellipsoid:', pigs.crs.ellipsoid)
print('Datum:', pigs.crs.datum)
print('Is geographic?:', pigs.crs.is_geographic)
print('Is projected?:', pigs.crs.is_projected)
```

```python
# Obtain the geographic extent of the geo-dataframe in long and lat
pigs.total_bounds
```

```python
# Examine CRS and extent of CA boundary
print('CRS:', ca_boundary.crs, '\n')
print(ca_boundary.total_bounds)
```

```python
pigs.crs == ca_boundary.crs
```

## Data wrangling

You can use all your `pandas` data wrangling to wrangle `geopandas.geoDataFrame`

```python
# Examine pig observations by year
pigs['year'].value_counts().sort_index()
```

```python
# Select data from 2020 onwards
pigs_recent = pigs[pigs['year']>=2020]

# Check length of original df
print("Total number of observations:", len(pigs))

# Check length of subset
print("Number of observations since 2020:", len(pigs_recent))
```

`gdp.GeoDataFrame` has a plot() method that quickly creates a mpa for our data using geospatial operations

## `matplotlib`

We can interact with `matplotlib` via its `pyplot` interface.

Matplotlib graphs the data in a **figure** that can have one or more **axes**. The axes is only the area specified bu the x axis and the y axis what is plotted in it. A figure can include multiple axes in it.

To create a blank figure:

```python
# Initialize empty figure (fig) and axis (ax)
fig, ax = plt.subplots() 

# Display figure
plt.show()
```

### Adding a layer

When using `matplotlib` it can be useful to think of creating a plot as adding layers on an axis.

Example: 
The first layer we want to add our axis is the `pigs_recent` point data

```python
# Initialize empty figure and axis
fig, ax = plt.subplots()

# Add pig point plot to our figure's axis 
pigs_recent.plot(ax = ax)

# Display figure
plt.show()
```

## Customization

Some can be done in `plot()` method for dataframe, others by updating attributes of the `ax` axis. 


```python
# Initialize empty figure and axis
fig, ax = plt.subplots()

# Add pig point plot to our figure's axis 
pigs_recent.plot(ax = ax,
                 alpha = 0.5,
                 color = "brown")

# Update axis
ax.set_title('Reported "Sus scrofa" sightings in CA (2020-2023)')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Display figure
plt.show()
```

```python
# Check-in

# Initialize empty figure and axis
fig, ax = plt.subplots()

# Add california boundary to our figure's axis 
ca_boundary.plot(ax = ax,
                 alpha = 0.5,
                 color = "none",
                 edgecolor = "hotpink")

# Add points of pig sightings
pigs_recent.plot(ax = ax,
                 alpha = 0.5,
                 color = "#FF5768",
                 edgecolor = "yellow")

# Update axis
ax.set_title('Reported "Sus scrofa" sightings in CA (2020-2023)')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
```
